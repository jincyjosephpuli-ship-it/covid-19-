### 2.1: Handling missing values ###

Handling missing values is one of the most important parts of data cleaning, especially for a project dealing with "Unspecified Countries" where missing or ambiguous data is the core problem

**Step 1:**  Find the Missing Values
Before fixing anything,  need to see where the holes in your data are. Run this code to get a count of missing values for each column:

```python

print(df.isnull().sum())
```



**Step 2:** Choose the Strategy
Depending on what column is missing data,  the three main ways to handle it:

**Strategy A:** Fill with a Placeholder (Best for the "Country" Column)
the country name is completely missing (blank or NaN),so don't want to throw that data away, because the whole point of the project is to analyze "Unspecified" trade! so it should fill those blanks with the label "Unspecified".

```python

df['Country'] = df['Country Name (country_name)'].fillna('Unspecified')
```

**Strategy B:** Impute with Mean or Median (Best for Numeric Values like Import_Value)

For a few missing numbers in the Quantity columns, deleting the whole row might lose valuable data. Instead, it can fill the gap using the Median (middle value) of that specific product category.

Using the median is safer than the average (mean) because trade data often has massive spikes (outliers) that can skew the average.

```python

median_value = df['Quantity of commodity (value_qt)'].median()
df['Quantity of commodity (value_qt)'] = df['Quantity of commodity (value_qt)'].fillna(median_value)
```

**Fill the Missing Codes with a Placeholder**

500 records represent  "Unspecified" data, it don't want to delete . Instead,  fill the missing codes with a clear placeholder text like "UNS" (for Unspecified) or 0.

```python

df['ISO Alpha 3 Code (alpha_3_code)'] = df['ISO Alpha 3 Code (alpha_3_code)'].fillna('UNS')

# Fill missing numeric codes with 0
df['Country Code (country_code) '] = df['Country Code (country_code)'].fillna(0)
df['Region Code (region_code)  '] = df['Region Code (region_code)'].fillna(0)
df['Sub-Region Code (sub_region_code) '] = df['Sub-Region Code (sub_region_code)'].fillna(0)

# Verify that all missing values are gone
print(df.isnull().sum())
```

## 2.4 Creating cerived Columns

```python
print(df.columns.tolist())
```

```python
import pandas as pd
import numpy as np

# 1. Rename the long columns to the short names we want to use
df = df.rename(columns={
    'Date (date)': 'date',
    'Value of commodity quantity in US Dollars (value_dl)': 'value_dl',
    'Quantity of commodity (value_qt)': 'value_qt',
    'Country Name (country_name)': 'country_name',
    'Harmonized System Code (hs_code)': 'hs_code'
})

# 2. Make sure 'date' is converted to proper datetime format
df['date'] = pd.to_datetime(df['date'])

# 3. Create your time-based derived columns
df['year'] = df['date'].dt.year
df['month'] = df['date'].dt.month
df['quarter'] = df['date'].dt.quarter

# 4. Create the 'is_unspecified' flag column
df['is_unspecified'] = df['country_name'].str.strip().str.lower().str.contains('unspecified|unknown|not specified', na=False)

# 5. Create the 'unit_price_usd' derived column (with safety check for 0 quantity)
df['unit_price_usd'] = np.where(
    df['value_qt'] > 0,
    df['value_dl'] / df['value_qt'],
    0.0
)

# 6. Create the 'hs_chapter' column
df['hs_chapter'] = df['hs_code'].astype(str).str.zfill(6).str[:2]

print("Success! Columns renamed and derived columns created without errors.")
print("\nYour confirmed active columns are now:")
print(df[['date', 'value_dl', 'year', 'is_unspecified', 'unit_price_usd']].head(3))
```

### Pivot table analysis

Pivot tables are one of the most powerful tools in data analysis. They allow you to cross-tabulate your data, instantly turning a flat table into a multidimensional grid (just like Excel pivot tables, but using Pandas!).

**Pivot Table 1:** Year vs. Unspecified Status (Total USD Value)

This table compares the total import value of Known vs. Unspecified countries side-by-side across different years. It’s perfect for spotting if unspecified trade is growing over time.

```python
# Pivot Table: Rows = Year, Columns = Unspecified Status, Value = Sum of USD
import_trend_pivot = df.pivot_table(
    index='year',
    columns='is_unspecified',
    values='value_dl',
    aggfunc='sum',
    margins=True,          # Adds a 'Total' row and column
    margins_name='Total'
)

# Format the numbers with commas and currency signs for better readability
pd.options.display.float_format = '${:,.2f}'.format
print("--- Total Import Value (USD) by Year and Unspecified Status ---")
print(import_trend_pivot)
```

### Corelation Analysis


Correlation analysis is a statistical method used to evaluate the strength and direction of the relationship between two or more quantitative variables.

In this trade dataset, correlation helps us answer critical questions like:

Does a higher quantity of imported goods (value_qt) always correlate with a higher dollar value (value_dl)?

**Understanding Correlation Values (r)**

Correlation is measured by a coefficient (usually Pearson's $r$) that ranges from -1.0 to +1.0:

$+1.0$ (Perfect Positive): As Variable A goes up, Variable B goes up in perfect proportion.

$0.0$ (No Correlation): There is no linear relationship between the variables (e.g., shipment date and country codes).

$-1.0$ (Perfect Negative): As Variable A goes up, Variable B goes down in perfect proportion.

**Step 1:** Run the Correlation Matrix in Pandas

### Statistical Summary

A statistical summary (often called descriptive statistics) is the first major step in quantitative analysis. It uses a core set of mathematical metrics to describe the central tendency, dispersion, and shape of your dataset’s distribution.

**1. The Core Metrics of a Statistical Summary**

To generate a statistical summary,that are looking at several distinct statistical concepts:

* Measures of Central Tendency: * Mean (Average): The sum of all values divided by the total number of values. It is highly sensitive to extreme outliers.

* Median (50th Percentile): The middle value when the data is sorted. It is a robust measure that isn't dragged away by massive outliers.

* Measures of Dispersion (Spread):

* Standard Deviation (Std): Measures how spread out your numbers are from the average. A low standard deviation means data points are clustered close to the mean; a high standard deviation means they are widely spread.

* Min & Max: The absolute smallest and largest values in the column.

* Percentiles (25%, 50%, 75%):

Help you understand the spread of your data. For example, if the 75th percentile of your shipment value is $10,000, it means 75% of your transactions are worth $10,000 or less.


**2. Generating the Summary in Python**
    
In Pandas,  generate a complete statistical summary for all numeric columns using a single, incredibly powerful function: df.describe().

```python
# 1. Get the statistical summary for all numeric columns
summary = df[['value_dl', 'value_qt', 'unit_price_usd']].describe()

# 2. Format the numbers with commas and two decimals so they are easy to read
formatted_summary = summary.apply(lambda x: x.apply(lambda val: f"{val:,.2f}"))

print("--- Cleaned Dataset Statistical Summary ---")
print(formatted_summary)
```

### 1. Line Chart: Yearly Import Trends (Known vs. Unspecified)

Why use it: This shows the trajectory of the data over time. It answers the crucial question: Is the total dollar value of unspecified trade growing, shrinking, or staying steady year-over-year?

### Donut Chart: Regional Composition of Unspecified Trade

Why use it: A donut chart (a pie chart with a clean cut-out center) is perfect for showing composition. It breaks down the percentage share that each global region (Asia, Europe, Americas, etc.) represents within the unspecified data bucket

```python
# 1. Get the value breakdown by region for unspecified data
region_shares = df[df['is_unspecified'] == True].groupby('Region Name (region)')['value_dl'].sum()

# 2. Plot the chart
plt.figure(figsize=(6, 6))
colors = sns.color_palette('pastel')[0:len(region_shares)]

plt.pie(
    region_shares,
    labels=region_shares.index,
    autopct='%1.1f%%',
    startangle=90,
    colors=colors,
    pctdistance=0.85,
    wedgeprops={'edgecolor': 'white', 'linewidth': 2} # Creates the clean panel dividers
)

# 3. Add the center circle to transform the pie chart into a modern donut chart
center_circle = plt.Circle((0,0), 0.70, fc='white')
fig = plt.gcf()
fig.gca().add_artist(center_circle)

plt.title('Regional Composition of Unspecified Trade Value', fontsize=14, pad=15)
plt.tight_layout()
plt.show()
```
